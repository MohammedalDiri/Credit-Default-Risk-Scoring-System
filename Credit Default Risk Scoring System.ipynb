```python
numerical_features_set = set(numerical_features)
def helper(df: pd.DataFrame, cols: list):
    rows = []

    for col in cols:
        missing_pct = df[col].isna().mean()


        rates = df.groupby(df[col].isna())['TARGET'].mean()
        default_present = rates.get(False, np.nan)
        default_missing = rates.get(True, np.nan)


        if col in numerical_features_set:
            corr_target = df[col].corr(df['TARGET'])
            n_unique = np.nan
        else:
            corr_target = np.nan
            n_unique = df[col].nunique()

        rows.append({
            'column': col,
            'missing_pct': missing_pct,
            'default_rate_present': default_present,
            'default_rate_missing': default_missing,
            'corr_with_target': corr_target,
            'n_unique': n_unique
        })

    return pd.DataFrame(rows)
```

```python
def aggregate_file(df, group_by_col, categorical_overrides=None, cols_to_exclude=None ,id_col=None, prefix=None):
  numerical_features1 = set(df.select_dtypes(include=[np.number]).columns)
  categorical_features1 = set(df.select_dtypes(include=['object']).columns)
  categorical_overrides = set(categorical_overrides or [])
  cols_to_exclude = set(cols_to_exclude or [])
  categorical_features1 = categorical_features1.union(categorical_overrides)
  numerical_features1 = numerical_features1.difference(categorical_overrides)
  numerical_features1 = numerical_features1.difference(cols_to_exclude)
  categorical_features1 = categorical_features1.difference(cols_to_exclude)
  numerical_features1 = numerical_features1.difference({group_by_col})
  numerical_features1 = numerical_features1.difference({id_col})
  categorical_features1 = categorical_features1.difference({group_by_col})
  dummies = pd.DataFrame(index=df.index)
  if categorical_features1:
    dummies = pd.get_dummies(df[list(categorical_features1)], dtype=bool)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(columns=list(categorical_features1))
  aggregations = {}
  if id_col :
    aggregations[id_col] = ["count"]
  for col in sorted(numerical_features1):
    aggregations[col] = ["mean", "max", "sum","min"]

  for col in sorted(dummies.columns):
    aggregations[col] = ["sum"]
  agg_df = df.groupby(group_by_col).agg(aggregations)
  agg_df.columns = [f"{col}_{agg}" for col, agg in agg_df.columns]
  if prefix:
    agg_df.columns = [f"{prefix}_{col}" for col in agg_df.columns]
  return agg_df

```
